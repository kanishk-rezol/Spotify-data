```python
import spotipy
import pandas as pd
import matplotlib.pyplot as plt
import re
from spotipy.oauth2 import SpotifyClientCredentials
import sqlite3
```

```python
conn=sqlite3.connect('example.db')
cursor=conn.cursor()
```

```python
sp= spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id="bfd9b8de09ac4f22864e9582f382f470",
    client_secret="26c61519aa8b48d6b01ada00bb75ebdc"
))
cursor.execute('''
CREATE TABLE IF NOT EXISTS new_table (
    ID INTEGER PRIMARY KEY AUTOINCREMENT,
    TrackName TEXT,
    Artist TEXT,
    Album TEXT,
    Popularity INTEGER,
    Duration REAL,
    Year INTEGER,
    ImageURL TEXT
)
''')
total_track=pd.read_csv(r"/content/dataset.csv")
for index, row in total_track.iterrows():
    track_id = str(row.get('track_id')).strip()
    if pd.isnull(track_id):
        continue
    try:
        track_url = f"https://open.spotify.com/track/{track_id}"
        print(f"Fetching: {track_url}")
        track = sp.track(track_id)
        track_name = track['name']
        artist = track['artists'][0]['name']
        album = track['album']['name']
        popularity = track['popularity']
        duration = track['duration_ms'] / 60000
        release_date = track['album']['release_date']
        year = int(release_date.split("-")[0])
        image_url = track['album']['images'][0]['url'] if track['album']['images'] else None
        cursor.execute('''
            INSERT INTO new_table (TrackName, Artist, Album, Popularity, Duration, Year, ImageURL)
            VALUES (?, ?, ?, ?, ?, ?, ?)
        ''', (track_name, artist, album, popularity, duration, year, image_url))
        conn.commit()

    except Exception as e:
        print(f"Failed to process track ID {track_id}: {e}")
# print(track)
```

```python
df = pd.read_sql("SELECT * FROM new_table", conn)
```
